```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv(r"Z:\Optics Temperature\Input Collimator.csv", delimiter=',')

columns = df.columns
print(df)

# Scatter plot for each time point
plt.scatter(df['amplification (A)'], df['Temperature (0mins)'], color='blue', label='Collimator (0mins)')
plt.scatter(df['amplification (A)'], df['Temperature (3mins)'], color='green', label='Collimator (3mins)')
plt.scatter(df['amplification (A)'], df['Temperature (6mins)'], color='red', label='Collimator (6mins)')

# Labeling
plt.xlabel('Amplification (A)')
plt.ylabel('Temperature (°C)')
plt.legend()

df2 = pd.read_csv(r"Z:\Optics Temperature\Amplifier Output.csv", delimiter=',')
columns2 = df2.columns

# Scatter plot for each time point
plt.scatter(df2['amplification (A)'], df2['Temperature (0mins)'], color='blue', label='Amplifier (0mins)', marker="s")
plt.scatter(df2['amplification (A)'], df2['Temperature (3mins)'], color='green', label='Amplifier (3mins)',  marker="s")
plt.scatter(df2['amplification (A)'], df2['Temperature (6mins)'], color='red', label='Amplifier (6mins)',  marker="s")

# Labeling
plt.xlabel('Amplification (A)')
plt.ylabel('Temperature (°C)')
plt.legend()

# Show plot
plt.show()

```

```python
df = pd.read_csv(r"Z:\Optics Temperature\25W 30mins.csv", delimiter=',')

columns = df.columns

fig, ax1 = plt.subplots(figsize=(10, 6))

# Plot temperature on the first y-axis
ax1.plot(df['time (mins)'], df['temperature'], color='red', marker='o', label='Temperature')
ax1.set_xlabel('Time (mins)')
ax1.set_ylabel('Temperature (°C)', color='red')
ax1.tick_params(axis='y', labelcolor='red')

# Create a second y-axis sharing the same x-axis
ax2 = ax1.twinx()
ax2.plot(df['time (mins)'], df['output (W)'], color='blue', marker='s', label='Output Power (W)')
ax2.set_ylabel('Output Power (W)', color='blue')
ax2.tick_params(axis='y', labelcolor='blue')

# Add a title
plt.title('25W')
# Show the plot
plt.show()
```
