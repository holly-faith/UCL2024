Plot Viscosity Force due to gas particles against particle velocity <br> <br>
$F_{viscosity} = 6 \pi \eta R v$ (source:Wikipedia)<br> <br>
where $\eta$ is the dynamic viscosity of air at atmosperic pressure, $R$ is the radius of the nanoparticle and $v$ is the flow velocity of the fluid relative to the particle (so just the particle's velocity if the fluid is stationary around it). <br> <br>
During loading, the particle is driven by 300mW in one direction, which corresponds to a (rough, didnt solve any integrals lol) force of about 50pN and accelerates until the viscous force due to the gas becomes equal to this. <br><br>
When the conveyor belt is turned on, the radiation force is (in theory) equal on both sides so the forces acting on it now consist of the viscous force and the confinement to the trapping site which pulls it forward. I'm not sure how much force the moving trapping site exerts as it moves (maybe it's in David's thesis) but if the trapping sites move faster than the particle can then maybe it gets knocked out and pushed along and we see more of a time averaged effect.
```python
import numpy as np
import matplotlib.pyplot as plt

eta_air = 18.5e-6 #Pascal seconds, standard atmospheric conditions 25deg and 1atm (source:Wikipedia)

def F_vis(v,R,eta):
    """
    input: 
    v in mm s-1 (conv to m s-1)
    R in nm (conv to m)
    eta in Pa s 

    output:
    F in pN (conv from N)
    """
    v = v*1e-3 
    R = R*1e-9
    
    F = 6*np.pi*eta*R*v
    F = F*1e12

    return F

v = np.linspace(0,5,100) # mm/s
R = 147.5 # nm

F_v = F_vis(v,R,eta_air)

plt.plot(v, F_v)
plt.xlabel('velocity (m/s)')
plt.ylabel('viscous force (pN)')
#plt.axhline(50)
plt.show()



```
