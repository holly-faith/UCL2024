```python
import os
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
# folder_path = r"Z:\Split Difference Photodiode Measurements\24-09-2024 Coherent Laser\before_loading_both_AOMs_spec"

# csv_files = [file for file in os.listdir(folder_path) if file.endswith('.csv')]

# csv_file = csv_files[0]

file_path_n = r"Z:\Split Difference Photodiode Measurements\24-09-2024 Coherent Laser\before_loading_both_AOMs_spec\before_loading_both_AOMs_spec_1.csv" #os.path.join(folder_path, csv_file) 
df_n = pd.read_csv(file_path_n)
frequency_n = np.asarray(df_n.iloc[:, 0],dtype = float) # first column
PSD_n = np.asarray(df_n.iloc[:, 1],dtype = float) # second column

file_path_p = r"Z:\Split Difference Photodiode Measurements\24-09-2024 Coherent Laser\1p_24-09_spec\1p_24-09_spec_1.csv"
df_p = pd.read_csv(file_path_p)
frequency_p = np.asarray(df_p.iloc[:, 0],dtype = float) # first column
PSD_p = np.asarray(df_p.iloc[:, 1],dtype = float) # second column



```

```python
plt.plot(frequency_n, PSD_n, label='no particle')
plt.plot(frequency_p, PSD_p, alpha=0.5, label='particle')
plt.xlim(0,100)
plt.xlabel('Frequency (KHz)')
plt.ylabel('PSD (dbm)')
plt.legend()
plt.show()
```
